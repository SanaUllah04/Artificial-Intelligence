##### 1. Logging level of the Transformers library to show only error messages, hiding warnings and info logs.

```python
from transformers.utils import logging

logging.set_verbosity_error()
```

<br>
<br>
<br>

##### 2. A text-to-speech pipeline using the model kakao-enterprise/vits-ljs to convert text into speech (audio output).

```python
from transformers import pipeline

narrator = pipeline(task="text-to-speech", model="kakao-enterprise/vits-ljs")
```

<br>
<br>
<br>

##### 3. Your Text

```python
text = """ 
Take a deep breath, and let yourself unwind. 
Imagine standing at the edge of a quiet lake, the water still and reflecting the sky above. 
A soft breeze brushes your skin, carrying the scent of fresh pine trees. 
Birds sing gently in the distance, their melodies calming and light. 
With each breath, feel your body relax, releasing all tension. 
You are safe, you are present, and you are at peace. Let the stillness embrace you. 
Let the quiet moments guide you back to yourself. In this place, you can simply be. 
You are enough, just as you are.
"""
```

<br>
<br>
<br>

##### 4. Converting the given **`text`** into speech audio using the loaded TTS pipeline and stores the result (audio data) in **`narrated_text`**.

```python
narrated_text = narrator(text)
```

<br>
<br>
<br>

##### 5. This plays the generated speech, using the audio data and its sampling rate from narrated_text.

```python
from IPython.display import Audio as ad

ad(narrated_text["audio"][0], rate=narrated_text["sampling_rate"])
```
